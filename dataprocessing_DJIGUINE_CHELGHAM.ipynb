```python
import pandas as pd
```

```python
ath_bio = pd.read_csv("/content/Olympic_Athlete_Bio.csv")
ath_event_result =pd.read_csv("/content/Olympic_Athlete_Event_Results.csv")
medal=pd.read_csv("/content/Olympic_Games_Medal_Tally.csv")
results=pd.read_csv("/content/Olympic_Results.csv")
country=pd.read_csv("/content/Olympics_Country.csv")
games=pd.read_csv("/content/Olympics_Games.csv")
#tweets=pd.read_csv("/content/tokyo_2020_tweets.csv")
```

```python
print(ath_bio.columns)
print(ath_event_result.columns)
print(medal.columns)
print(results.columns)
print(country.columns)
print(games.columns)
#print(tweets.columns)
```

```python
ath_bio['athlete_id'] = ath_bio['athlete_id'].astype(str) +'a'
ath_event_result['athlete_id'] = ath_event_result['athlete_id'].astype(str) +'a'
ath_event_result['edition_id'] = ath_event_result['edition_id'].astype(str) +'e'
games['edition_id'] = games['edition_id'].astype(str) +'e'
results['edition_id'] = results['edition_id'].astype(str) +'e'
medal['edition_id'] = medal['edition_id'].astype(str) +'e'
```

**Traitement des données pour prendre en comptes les contraintes speciée**

**LEFT JOIN Olympic_Athlete_Event_Results.csv file with Olympic_Athlete_Bio.csv**

```python
print("taille de athlete bio =",ath_bio.size)
print("taille de event_athlete_result =",ath_event_result.size)
ath_event_result['athlete'] = ath_event_result.groupby('athlete_id')['athlete'].transform('last')
ath_bio['name'] = ath_bio.groupby('athlete_id')['name'].transform('last')
```

```python
merged_ath_bio_and_ath_event_result = pd.merge(ath_bio, ath_event_result, on='athlete_id', how='left')
merged_ath_bio_and_ath_event_result['name'] = merged_ath_bio_and_ath_event_result.groupby('athlete_id')['name'].transform('last')
print(merged_ath_bio_and_ath_event_result.size)
merged_ath_bio_and_ath_event_result
```

**LEFT JOIN Olympic_Athlete_Event_Results.csv with Olympic_Results.csv **

```python
merged_ath_event_result_and_results = pd.merge(ath_event_result, results, on='result_id', how='left')

print(merged_ath_event_result_and_results.size)
merged_ath_event_result_and_results.head()
```

**Construction du fichier csv  des athletes**



```python
athlete_columns = ['athlete_id',	'name',	'sex',	'born',	'height',	'weight',	'country',	'country_noc',]
athlete = merged_ath_bio_and_ath_event_result[['athlete_id',	'name',	'sex',	'born',	'height',	'weight',	'country',	'country_noc_x',	'description',	'special_notes']]
athlete = athlete.rename(columns={'country_noc_x': 'country_noc'})
athlete = athlete[athlete_columns].drop_duplicates()
athlete['label'] = 'athlete'


athlete.to_csv('athlete.csv', index=False)
# Création du fichier d'en-tête
with open('athlete_header.csv', 'w') as f:
    f.write('athlete_id:ID,name,sex,born,height,weight,country,country_noc,description,special_notes,:LABEL\n')

print(athlete.size)
athlete
```

```python
# Extraire les athlètes uniques à partir des résultats d'événements
athletes_from_events = ath_event_result[['athlete_id', 'athlete']].drop_duplicates()

# Renommer les colonnes pour correspondre au fichier de bio
athletes_from_events.columns = ['athlete_id', 'name']

# Supprimer les doublons dans ath_bio (si nécessaire)
ath_bio = ath_bio.drop_duplicates(subset=['athlete_id'])

# Fusionner avec les informations biographiques
athletes_df = pd.merge(athletes_from_events, ath_bio, on='athlete_id', how='left')
athletes_df = athletes_df.drop_duplicates(subset=['athlete_id'])
athletes_df['label'] = 'athlete'
athletes_df = athletes_df.drop('name_y', axis=1)
# Remplacer les valeurs manquantes par "UNKNOWN" ou une valeur par défaut
athletes_df.fillna({
    'sex': 'UNKNOWN',
    'born': 'UNKNOWN',
    'height': 'UNKNOWN',
    'weight': 'UNKNOWN',
    'country': 'UNKNOWN',
    'country_noc': 'UNKNOWN',
    'description': 'UNKNOWN',
    'special_notes': 'UNKNOWN'
}, inplace=True)

# Supprimer les doublons après la fusion (au cas où)
athletes_df = athletes_df.drop_duplicates(subset=['athlete_id'])
# Sauvegarder le fichier athletes.csv
athletes_df.to_csv('athlete.csv', index=False)

# Créer le fichier athletes_header.csv
with open('athlete_header.csv', 'w') as f:
    f.write('athlete_id:ID,name,sex,born,height,weight,country,country_noc,description,special_notes,:LABEL\n')

athletes_df
```

```python
edition = pd.read_csv("/content/edition.csv")

# Renommer la colonne noc pour aligner avec les autres tables si nécessaire
country = country.rename(columns={'noc': 'country_noc'})

# Joindre les résultats des athlètes avec les éditions
medals_with_editions = pd.merge(ath_event_result, edition[['edition_id', 'edition', 'year']], on='edition_id', how='inner')
medals_with_editions  = medals_with_editions .rename(columns={'edition_x': 'edition'})
medals_with_editions  = medals_with_editions .rename(columns={'event': 'discipline'})

# Grouper et compter les médailles par pays, édition et type
medal_counts_with_editions = medals_with_editions.groupby(['country_noc','medal', 'edition_id', 'edition', 'year','sport','discipline']).size().reset_index(name='count')
# Supprimer les lignes avec count = 0 si nécessaire
medal_counts_with_editions = medal_counts_with_editions[medal_counts_with_editions['count'] > 0]

# Ajouter la colonne pour le type de relation
medal_counts_with_editions[':TYPE'] = 'A_obtenu'
columns = medal_counts_with_editions.columns.tolist()  # Obtenir la liste des colonnes
medal_index = columns.index('medal')  # Trouver l'index de la colonne 'medal'
columns.insert(medal_index + 1, columns.pop(columns.index(':TYPE')))  # Déplacer ':TYPE' après 'medal'
medal_counts_with_editions = medal_counts_with_editions[columns]

# Exporter le fichier CSV
medal_counts_with_editions.to_csv('A_obtenu.csv', index=False)

# Écrire un en-tête personnalisé
with open('A_obtenu_header.csv', 'w') as f:
    f.write(':START_ID,:END_ID,:TYPE,edition_id,edition,year,count,sport,discipline')

medal_counts_with_editions.head()
```
